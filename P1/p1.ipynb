### 1.4 Images manipulation

a) Add a black frame to Lenna's image. It should have a width of 1/10th of the height of the original image. We recommend to create a bigger black image to later embed the original image in the central part.

```python
def create_frame(image,width):#image: image to modify , width: desired width for the frame
                             #return framed image
    if len(image.shape) == 3:#we have value as rgb [R,G,B]
        #we create an image with the extra frames using the slicing the images should be (lena + 2width) in both axis
        frame = np.zeros((image.shape[0]+(2*width),image.shape[1]+(2*width),3),dtype='uint8') 
    else:#we have grayscaled image just 1 value
        frame = np.zeros((image.shape[0]+(2*width),image.shape[1]+(2*width)),dtype='uint8')
    frame[width:frame.shape[0]-width,width:frame.shape[1]-width] = image#we insert the image inside the frame
    return frame 
    
im = io.imread('images/lenna.jpg')
frame_width = int(im.shape[0]/10)

frame = create_frame(im,frame_width)

plt.imshow(frame)
```

b) Change the mask of the framed image to red. Visualize the three images with their respective titles.

```python
def changeColorFrame(frame,color):#frame: image with the frame ,color: color to fill the frame
                                  #return: colored frame image
    #slice every section of the frame and filling with the color
    frame[:,:frame_width] = color
    frame[:,frame.shape[1]-frame_width:frame.shape[1]] = color
    frame[:frame_width,:] = color
    frame[frame.shape[0]-frame_width:frame.shape[0],:] = color
    return frame
frame = changeColorFrame(frame,[255,0,0])
plt.imshow(frame)
```

```python
#Original Lena
im = io.imread('images/lenna.jpg')
frame_width = int(im.shape[0]/10)
im = create_frame(im,frame_width)
im = changeColorFrame(im,[255,0,0])
#Grey Lena
im_g = io.imread('images/lenna_g.jpg')
frame_width = int(im_g.shape[0]/10)
im_g = create_frame(im_g,frame_width)
#Bin Lena
im_bin = io.imread('images/lenna_bin.jpg')
frame_width = int(im_bin.shape[0]/10)
im_bin = create_frame(im_bin,frame_width)
#Displaying
plt.title('Original')
plt.imshow(im)
plt.show()
plt.title('Grayscaled')
plt.imshow(im_g, cmap='gray')
plt.show()
plt.title('Binarized')
plt.imshow(im_bin, cmap='binary')
plt.show()
```

###Â 1.5 Contrast enhancement

a) Open the image `car_lowContrast.jpg` and visualize it without graylevel enhancement. To this purpose, check which are the minimal and maximal values and how to use the command imshow without contrast enhancement.

Note: When running the function `imshow()` the image is authomatically enhanced. Use the `vmin` and `vmax` parameters in order to deactivate the enhancement applied by default.

```python
car = io.imread('images/car_lowContrast.jpg')#Original Car
plt.imshow(car,cmap='gray',vmax=255,vmin=0)
```

b) Create an image called `dark_image` using the  values from the original one, so that the minimal value of the new image is 0. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
#((image  - imagemin)/(imagemax - imagemin)) * (newscalemax - newscalemin) + newscalemin
dark_image = car.copy()
min_d = dark_image.min()
max_d = dark_image.max()
dark_image = ((dark_image - min_d)/(max_d - min_d)) * (max_d - 0) + 0 #applying reescale contrast formula
#Comparing values
print("Car max: "+str(car.max()))
print("Dark max: "+str(dark_image.max()))
print("Car min: "+str(car.min()))
print("Dark min: "+str(dark_image.min()))
```

c) Create a new image called `bright_im` from the original one, so that the maximal value of the new image is 255. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
#((image  - imagemin)/(imagemax - imagemin)) * (newscalemax - newscalemin) + newscalemin
bright_im = car.copy()
min_b = bright_im.min()
max_b = bright_im.max()
bright_im = ((bright_im - min_b)/(max_b - min_b)) * (255 - min_b) + min_b #again formula
#Comparing values
print("Car max: "+str(car.max()))
print("Bright max: "+str(bright_im.max()))
print("Car min: "+str(car.min()))
print("Bright min: "+str(bright_im.min()))
```

### 1.6 Masks and logical image manipulation

Read the image `circles.bmp` (see figure top) and define 3 images in such a way that each one contains only 1 ball (see figure bottom). Visualize the 3 images in the same figure, each one with its corresponding title.

<img src="images_notebook/ex16.png">

**Note:** Be careful when applying `imshow` since it automatically increases the contrast of the image (putting minimum values to 0 and maximal to 1).

**Note:** Be careful when applying type conversion. Check the ranges of the image values.

```python
#Black Circle
circles = io.imread('images/circles.bmp')
black = np.zeros(circles.shape)
black[:,:,:] = [255,255,255] #all image white
mask = circles[:,:] == np.array([0,0,0])#creating mask
black[mask] = circles[mask] #copying values chosen by the mask
black = black.astype(np.uint8)
plt.title('Left Black')
plt.imshow(black,vmax=255,vmin=0)
plt.show()
#Dark Grey Circle same process but different mask
d_grey = np.zeros(circles.shape)
d_grey[:,:,:] = [255,255,255]
mask = circles[:,:] == np.array([127,127,127])
d_grey[mask] = circles[mask]
d_grey = d_grey.astype(np.uint8)
plt.title('Center Dark Grey')
plt.imshow(d_grey,vmax=255,vmin=0)
plt.show()
#Bright Grey Circle same process but different mask
b_grey = np.zeros(circles.shape)
b_grey[:,:,:] = [255,255,255]
mask = circles[:,:] == np.array([195,195,195])
b_grey[mask] = circles[mask]
b_grey = b_grey.astype(np.uint8)
plt.title('Rigth Bright Grey')
plt.imshow(b_grey,vmax=255,vmin=0)
plt.show()

```

e) Given the images `coat.png` and `model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is only an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying resize, rotate, etc. keep track how the image type and values range change. When necessary, apply type conversion using img_as_float, img_as_uint, img_as_ubyte.

```python
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.color import rgba2rgb
#We load images, change all of them to RGB and pass to uint8
#texture additional is resized to the other's image shape
coat = io.imread('images/coat.png')
coat = rgba2rgb(coat)
coat *= 255
coat = coat.astype(np.uint8)
model = io.imread('images/model.png')
model = rgba2rgb(model)
model *= 255
model = model.astype(np.uint8)
texture = io.imread('images/texture.jpg')
texture = resize(texture, (1188, 915), anti_aliasing=True)
texture *= 255
texture = texture.astype(np.uint8)
#We use the coat image for detecting the pixels that would be use for the coat and replace them in the model image 
#for the ones in the texture
for i in range(coat.shape[0]):
    for j in range(coat.shape[1]):
        if not np.array_equal(coat[i,j],np.array([0,0,0])):
            model[i,j] = texture[i,j]
plt.title('Textured')
plt.imshow(model)
```
